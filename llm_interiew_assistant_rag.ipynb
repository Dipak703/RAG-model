```python
#Loading Data from pdf
from llama_index.core import SimpleDirectoryReader
data = SimpleDirectoryReader("D:\Beginner\dhiwise\data").load_data()
```

```python
#preprocessing of data
documents = []
metadata = []
ids = []
for items in data:
    items=dict(items)
    documents.append(items["text"])
    metadata.append(items["metadata"])
    ids.append(items["id_"])
```

```python
# import dependency for vector database and llm model
import chromadb
from langchain_community.llms import Ollama
```

```python
# initilizing the database and fetch data or registring data
chroma_client = chromadb.PersistentClient(path='D:\Beginner\dhiwise\chromabd')
collection = chroma_client.get_or_create_collection(name="dhiwise_ai_job")
```

```python
# uploading data to vector database
collection.upsert(
    documents=documents,
    metadatas=metadata,
    ids=ids
)
```

```python
# system prompt
system_prompt="""
You are an assistant. Your goal help and assist in the job hiring process.by Providing analytics to show which candidate is a good fit for a given job,
        highlighting the percentage of required skills that candidate possesses

"""
```

```python
# calling the model
llm = Ollama(model="llama2",base_url  = "http://localhost:11434",system=system_prompt,temperature=0.0)
```

```python
# function for resume analysics
def check_eligibility(user_query):
    results = collection.query(
        query_texts=[user_query],
        n_results=5
    )                       # fetching data from vectorembeddings using user_query
    prompt=f"""<|USER|>{user_query} , analyze the job requirement and different candidates data given to you. give how much this prencent candidate is suitable for this job role.
        job requirment :{results['documents'][0][0]},
        candidate 1 :{results['documents'][0][2]},
       , candidate 2 :{results['documents'][0][1]}
       <|ASSISTANT|>"""

    return llm.invoke(prompt) # giving prompt with vector data to llm
```

```python
# import dependency for transcribing
import assemblyai as aai
aai.settings.api_key = "35a89577b4804bbf9d1576442bb7621c"
```

```python
# for interview analysics
def analyse_interview(input):
    try:
        transcriber = aai.Transcriber()
        interview = transcriber.transcribe(input)

    except:
        print("failed")

    prompt1=f"""<|USER|>analyze the interview conversation of candidate and give information like what is interview about, how confident the candidate is, how accurate the candidate answers are etc.
      here is the interview conversation= {interview.text}

       <|ASSISTANT|>"""

    return llm.invoke(prompt1)
```

```python
#calling the model for eligibility checking
a=check_eligibility("check is the candidates are suitable for dhiwise AI job role and give how much percentage of skills a matching")
```
